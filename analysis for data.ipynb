```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import randint
```

```python
df=pd.read_csv(r"C:\Users\otaku7\Desktop\GP\clean_data.csv")
df.head()
```

```python
#correlation matrix
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
plt.show()
```

# treatment correlation matrix

```python
k = 11
cols = corrmat.nlargest(k, 'treatment')['treatment'].index
# np.corrcoef to compute the correlation coefficients between these columns.
corrm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(corrm, cbar=True, annot=True, square=True, fmt='.2f',
                 annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
# treatment and age
ta = sns.FacetGrid(df, col='treatment', size=5)
ta= ta.map(sns.distplot, "Age")
```
